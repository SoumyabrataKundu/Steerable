```python
import torch
import numpy as np
import matplotlib.pyplot as plt
from Steerable.utils import Reconstruct

```

```python
reconstruct = Reconstruct((256,256), (1000,1000), 64, sigma=50)
entropy = reconstruct(-torch.log(reconstruct.weights)) / np.log(reconstruct.num_patches)
entropy = torch.nan_to_num(entropy, 0)
plt.imshow(entropy)
```
